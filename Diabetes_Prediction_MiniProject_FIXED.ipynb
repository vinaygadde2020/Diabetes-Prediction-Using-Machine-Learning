## 📖 Section 3: Introduction

### 3.1 Introduction to Machine Learning
Machine Learning (ML) is a branch of Artificial Intelligence that enables computers to learn from data and make predictions or decisions without being explicitly programmed for each task. ML algorithms build mathematical models from training data to generalize patterns to unseen examples. The three main paradigms are:
- **Supervised Learning** – Model learns from labeled data (classification, regression).
- **Unsupervised Learning** – Model finds hidden structure in unlabeled data (clustering, dimensionality reduction).
- **Reinforcement Learning** – Agent learns through trial-and-error interactions with an environment.

### 3.2 Introduction to Classification
Classification is a supervised learning task where the model predicts a **discrete class label** for a given input. The model learns a decision boundary that separates different classes in the feature space. Examples include spam detection (spam / not spam), image recognition (cat / dog), and disease prediction (diabetic / not diabetic).

### 3.3 What is Diabetes?
Diabetes mellitus is a chronic disease that occurs when the body cannot produce or effectively use insulin — a hormone that regulates blood glucose. There are three main types:
- **Type 1 Diabetes** – The immune system destroys insulin-producing cells (autoimmune).
- **Type 2 Diabetes** – The body becomes resistant to insulin; most common form (~90% of cases).
- **Gestational Diabetes** – Develops during pregnancy and may persist after delivery.

According to the WHO, over **422 million people** globally have diabetes, and this number is rising rapidly, particularly in low- and middle-income countries.

## 🔧 Section 10: Data Preprocessing

### Step 1: Missing Value Analysis

```python
# ── Check for NaN missing values ─────────────────────────────────────────────
missing = df.isnull().sum()
print("MISSING VALUE ANALYSIS")
print("=" * 40)
print(missing)
print(f"\nTotal missing values: {missing.sum()}")
print("\n✅ No NaN values — but zeros may represent missing data in clinical features.")
```

**Explanation:** While no NaN values exist, the dataset encodes missing data as zeros for clinical measurements. A blood pressure or glucose of zero is physiologically impossible, so these must be treated as missing.

### Step 2: Check Duplicate Records

```python
# ── Duplicate check ───────────────────────────────────────────────────────────
duplicates = df.duplicated().sum()
print(f"Duplicate rows found: {duplicates}")

if duplicates > 0:
    df = df.drop_duplicates()
    print(f"✅ Removed. New shape: {df.shape}")
else:
    print("✅ No duplicate records found.")
```

### Step 3: Replace Invalid Zero Values with Median

```python
# ── Columns where 0 is medically impossible ───────────────────────────────────
zero_invalid_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

print("ZERO VALUE COUNT BEFORE REPLACEMENT")
print("=" * 50)
for col in zero_invalid_cols:
    n_zeros = (df[col] == 0).sum()
    print(f"  {col:<30}: {n_zeros} zeros ({n_zeros/len(df)*100:.1f}%)")

# ── Replace zeros with median (computed on non-zero values only) ───────────────
df_clean = df.copy()
for col in zero_invalid_cols:
    median_val = df_clean[col][df_clean[col] != 0].median()
    df_clean[col] = df_clean[col].replace(0, median_val)
    print(f"  ✅ {col}: zeros replaced with median = {median_val:.2f}")

print(f"\nDataset shape after cleaning: {df_clean.shape}")
```

**Explanation:** We replace zero values with the **median** of non-zero values (rather than the mean) because the median is robust to outliers that may skew the distribution. For example, `Insulin` has 374 zeros (48.7%) — replacing these with the median of valid readings preserves the distributional integrity of the feature.

## 🤖 Section 12: Machine Learning Models

```python
# ── Evaluation helper function ────────────────────────────────────────────────
all_results = []
all_predictions = {}
all_probabilities = {}

def evaluate_classifier(name, model, X_tr, X_te, y_tr, y_te):
    """Train classifier and compute all evaluation metrics."""
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    y_prob = model.predict_proba(X_te)[:, 1] if hasattr(model, 'predict_proba') else None

    acc   = accuracy_score(y_te, y_pred)
    prec  = precision_score(y_te, y_pred, zero_division=0)
    rec   = recall_score(y_te, y_pred, zero_division=0)
    f1    = f1_score(y_te, y_pred, zero_division=0)
    cv    = cross_val_score(model, X_tr, y_tr, cv=5, scoring='accuracy').mean()
    auc_s = roc_auc_score(y_te, y_prob) if y_prob is not None else None

    results = {'Model': name, 'Accuracy': acc, 'Precision': prec,
               'Recall': rec, 'F1 Score': f1, 'CV Accuracy': cv,
               'AUC': auc_s if auc_s else 'N/A'}

    all_predictions[name] = y_pred
    if y_prob is not None:
        all_probabilities[name] = y_prob

    print(f"{'='*55}")
    print(f"  {name}")
    print(f"{'='*55}")
    print(f"  Accuracy  : {acc:.4f} ({acc*100:.2f}%)")
    print(f"  Precision : {prec:.4f}")
    print(f"  Recall    : {rec:.4f}")
    print(f"  F1 Score  : {f1:.4f}")
    print(f"  CV Acc    : {cv:.4f}")
    if auc_s: print(f"  AUC       : {auc_s:.4f}")
    return results, model

print("✅ Evaluation function defined.")
```

### 12.1 Logistic Regression

**Theory:** Logistic Regression models the probability of a binary outcome using the **sigmoid function**:

$$P(y=1 | \mathbf{x}) = \sigma(\mathbf{w}^T \mathbf{x} + b) = \frac{1}{1 + e^{-(\mathbf{w}^T \mathbf{x} + b)}}$$

The model is trained by minimizing the **binary cross-entropy loss**:

$$\mathcal{L} = -\frac{1}{n} \sum_{i=1}^{n} \left[ y_i \log(\hat{p}_i) + (1-y_i) \log(1-\hat{p}_i) \right]$$

**Advantages:** Interpretable coefficients, fast training, provides probability scores, works well when classes are linearly separable.  
**Disadvantages:** Assumes linear decision boundary; poor performance on complex non-linear relationships.


```python
# ── Logistic Regression ───────────────────────────────────────────────────────
lr = LogisticRegression(max_iter=1000, random_state=42)
res_lr, lr_model = evaluate_classifier(
    'Logistic Regression', lr, X_train_sc, X_test_sc, y_train, y_test
)
all_results.append(res_lr)
```

### 12.2 K-Nearest Neighbors (KNN)

**Theory:** KNN classifies a new data point based on the **majority class among its K nearest neighbors** in the feature space, using Euclidean distance:

$$d(\mathbf{x}, \mathbf{x'}) = \sqrt{\sum_{j=1}^{p} (x_j - x'_j)^2}$$

The class is assigned by a majority vote of the K neighbors. The choice of K is critical — small K overfits, large K underfits.

**Advantages:** Simple, non-parametric, naturally handles multi-class problems.  
**Disadvantages:** Slow prediction on large datasets (computes all distances), sensitive to feature scale and irrelevant features.


```python
# ── KNN Classifier ────────────────────────────────────────────────────────────
knn = KNeighborsClassifier(n_neighbors=5)
res_knn, knn_model = evaluate_classifier(
    'KNN', knn, X_train_sc, X_test_sc, y_train, y_test
)
all_results.append(res_knn)
```

### 12.3 Decision Tree Classifier

**Theory:** A Decision Tree recursively partitions the feature space by selecting the feature and threshold that maximizes **Information Gain** (or minimizes **Gini Impurity**):

$$\text{Gini}(t) = 1 - \sum_{k} p_k^2$$

$$\text{Entropy}(t) = -\sum_{k} p_k \log_2(p_k)$$

The split is chosen to maximize the reduction in impurity between the parent and child nodes.

**Advantages:** Interpretable, handles non-linear boundaries, no feature scaling needed.  
**Disadvantages:** High variance (overfitting), unstable — small data changes can produce very different trees.


```python
# ── Decision Tree Classifier ──────────────────────────────────────────────────
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
res_dt, dt_model = evaluate_classifier(
    'Decision Tree', dt, X_train, X_test, y_train, y_test
)
all_results.append(res_dt)
```

### 12.4 Random Forest Classifier

**Theory:** Random Forest is a **bagging** ensemble of Decision Trees. Each tree is trained on a random bootstrap sample, and each split considers only a random subset of features (`max_features`). The final prediction is the **majority vote** of all trees:

$$\hat{y} = \text{mode}\left(T_1(\mathbf{x}), T_2(\mathbf{x}), \ldots, T_B(\mathbf{x})\right)$$

This ensemble approach dramatically reduces the variance of individual trees while maintaining low bias.

**Advantages:** High accuracy, resistant to overfitting, built-in feature importance, handles class imbalance well.  
**Disadvantages:** Less interpretable than a single tree, computationally heavier, can be slow on very large datasets.


```python
# ── Random Forest Classifier ──────────────────────────────────────────────────
rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
res_rf, rf_model = evaluate_classifier(
    'Random Forest', rf, X_train, X_test, y_train, y_test
)
all_results.append(res_rf)
```

### 12.5 Support Vector Machine (SVM)

**Theory:** SVM finds the **optimal hyperplane** that maximizes the margin between the two classes:

$$\underset{\mathbf{w}, b}{\min} \frac{1}{2} \|\mathbf{w}\|^2 \quad \text{s.t.} \quad y_i(\mathbf{w}^T \mathbf{x}_i + b) \geq 1$$

The **RBF kernel** maps data to a higher-dimensional space to handle non-linear separability:

$$K(\mathbf{x}, \mathbf{x'}) = \exp\left(-\gamma \|\mathbf{x} - \mathbf{x'}\|^2\right)$$

**Advantages:** Effective in high-dimensional spaces, strong theoretical foundations, handles non-linear boundaries with kernels.  
**Disadvantages:** Does not scale well to very large datasets, sensitive to feature scaling, kernel choice impacts performance significantly.


```python
# ── Support Vector Machine ────────────────────────────────────────────────────
svm = SVC(kernel='rbf', probability=True, random_state=42)
res_svm, svm_model = evaluate_classifier(
    'SVM', svm, X_train_sc, X_test_sc, y_train, y_test
)
all_results.append(res_svm)
```

## 🏥 Section 17: Interactive Prediction System

```python
# ══════════════════════════════════════════════════════════════════════════════
# DIABETES PREDICTION SYSTEM
# ══════════════════════════════════════════════════════════════════════════════

def validate_input(name, value, min_val, max_val):
    """Validate that a clinical measurement is within a reasonable range."""
    if value < min_val or value > max_val:
        print(f"  ⚠️  Warning: {name} = {value} is outside typical range [{min_val}, {max_val}]")
    return value

def predict_diabetes(pregnancies, glucose, blood_pressure, skin_thickness,
                     insulin, bmi, diabetes_pedigree, age):
    """
    Predict diabetes risk for a patient.

    Parameters:
    -----------
    pregnancies       : int   – Number of pregnancies (0–17)
    glucose           : int   – Plasma glucose concentration (mg/dL, 44–199)
    blood_pressure    : int   – Diastolic blood pressure (mm Hg, 24–122)
    skin_thickness    : int   – Skin fold thickness (mm, 7–99)
    insulin           : int   – Serum insulin (mu U/ml, 14–846)
    bmi               : float – Body Mass Index (18.2–67.1)
    diabetes_pedigree : float – Diabetes pedigree function (0.08–2.42)
    age               : int   – Age in years (21–81)

    Returns:
    --------
    diagnosis, probability
    """
    # ── Input validation ───────────────────────────────────────────────────────
    validate_input('Pregnancies', pregnancies, 0, 17)
    validate_input('Glucose', glucose, 44, 199)
    validate_input('BloodPressure', blood_pressure, 24, 122)
    validate_input('SkinThickness', skin_thickness, 7, 99)
    validate_input('Insulin', insulin, 14, 846)
    validate_input('BMI', bmi, 18.2, 67.1)
    validate_input('DiabetesPedigreeFunction', diabetes_pedigree, 0.08, 2.42)
    validate_input('Age', age, 21, 81)

    # ── Build input array ──────────────────────────────────────────────────────
    input_data = np.array([[pregnancies, glucose, blood_pressure, skin_thickness,
                            insulin, bmi, diabetes_pedigree, age]])
    input_df = pd.DataFrame(input_data, columns=X.columns)

    # ── Random Forest doesn't require scaling; use unscaled input ──────────────
    prediction = final_model.predict(input_df)[0]
    probability = final_model.predict_proba(input_df)[0]

    return prediction, probability


# ══════════════════════════════════════════════════════════════════════════════
# TEST CASE 1 – High-Risk Patient
# ══════════════════════════════════════════════════════════════════════════════
pred, prob = predict_diabetes(
    pregnancies=6, glucose=148, blood_pressure=72,
    skin_thickness=35, insulin=0, bmi=33.6,
    diabetes_pedigree=0.627, age=50
)

print("╔══════════════════════════════════════════════════════════╗")
print("║              DIABETES RISK ASSESSMENT REPORT            ║")
print("╠══════════════════════════════════════════════════════════╣")
print(f"║  Pregnancies             : 6                           ║")
print(f"║  Glucose (mg/dL)         : 148                         ║")
print(f"║  Blood Pressure (mm Hg)  : 72                          ║")
print(f"║  Skin Thickness (mm)     : 35                          ║")
print(f"║  Insulin (mu U/ml)       : 0 (replaced by median)      ║")
print(f"║  BMI                     : 33.6                        ║")
print(f"║  Diabetes Pedigree Func  : 0.627                       ║")
print(f"║  Age                     : 50                          ║")
print("╠══════════════════════════════════════════════════════════╣")
if pred == 1:
    print(f"║  🔴 DIAGNOSIS  : DIABETIC                              ║")
else:
    print(f"║  🟢 DIAGNOSIS  : NOT DIABETIC                          ║")
print(f"║  Probability Not Diabetic: {prob[0]*100:.1f}%                       ║")
print(f"║  Probability Diabetic    : {prob[1]*100:.1f}%                        ║")
print(f"║  Confidence Level        : {'HIGH' if max(prob) > 0.75 else 'MODERATE'}                          ║")
print("╚══════════════════════════════════════════════════════════╝")


# ══════════════════════════════════════════════════════════════════════════════
# TEST CASE 2 – Low-Risk Patient
# ══════════════════════════════════════════════════════════════════════════════
print("\n")
pred2, prob2 = predict_diabetes(
    pregnancies=1, glucose=85, blood_pressure=66,
    skin_thickness=29, insulin=0, bmi=26.6,
    diabetes_pedigree=0.351, age=31
)
if pred2 == 1:
    diag2 = "🔴 DIABETIC"
else:
    diag2 = "🟢 NOT DIABETIC"

print(f"Test Case 2 – Low-Risk Patient:")
print(f"  Diagnosis       : {diag2}")
print(f"  P(Not Diabetic) : {prob2[0]*100:.1f}%")
print(f"  P(Diabetic)     : {prob2[1]*100:.1f}%")
```
